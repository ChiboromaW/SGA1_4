```python
# Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import library
import sqlite3

# Create connection to the database
conn = sqlite3.connect("attr_data.db")

#Create the cursor object
cursor = conn.cursor()
```

```python
# Extracting the data and importing it to pandas

df = pd.read_sql("SELECT * FROM 'attrition_records';", conn)

# To show the first 5 rows in the table
df.head()
```

```python
# Assessing the dataset to view the shape of the table, and table information 
print(df.shape)
print(df.info())
df.describe()
```

```python
# Assessing the table to chheck for any missing and duplicate values 
print(df.isna().sum())
print(df.duplicated())
```

From assessing the table, we notice that
1. The column datatype is of objects and intergers.
2. There is no missing value nor duplicated value
3. The datatype of each column is correct, hence no need for further cleaning.

**QUESTION** **1**

```python
# Plotting Heatmaps
plt.figure(figsize=[18,14])
sns.heatmap(df.corr(), annot = True, fmt = '.2f')
plt.title('IBM Employee Attribution Correlation Heatmap');
```

When we interpret the correlations in the above heatmap, we see that **Job Level**, **Monthly Income** and **Total Working Years** are the first 3 with highest values, hence they are the major factors of employee attrition.

**QUESTION 2**

```python
plt.figure(figsize=[8,5])
sns.boxplot(data=df, x='Education', y='MonthlyIncome', palette= 'Accent',fliersize = 4, width = 0.8)
plt.title('Boxplot for the relationship between Education and Monthly Income')
plt.xlabel('Education')
plt.ylabel('Monthly Income');
```

**EXPLANATION**: From the boxplot above, we notice that even with a high number of degrees, there is no much difference in the monthyl income, hence we can say that, the number of degrees does not affect the monthyl income.

**QUESTION 3**

```python
plt.figure(figsize=[8,5])
sns.boxplot(data=df, y='Age', x='Attrition', palette= 'Accent',fliersize = 4, width = 0.8)
plt.title('Boxplot for the effect of Age on Attrition')
plt.ylabel('Age')
plt.xlabel('Attrition');
```

**EXPLANATION:** From the boxplot above, we observe that majority of the employees who left the company fall between 28-40 years. Hence, age(aging) is not a major factor to employee attrition 

**QUESTION 4**

```python
plt.figure(figsize=[8,4])
plt.hist(df[df['Attrition']== 'Yes']['MonthlyIncome'], label ='Yes',alpha =0.5,density = True)
plt.hist(df[df['Attrition']== 'No']['MonthlyIncome'], label ='No',alpha =0.5,density = True)
plt.title('Histogram of Attrition In Respect to Monthly Income')
plt.xlabel('Montthly Income')
plt.ylabel('Counts')
plt.legend();
```

**EXPLANATION:** From the above plot, it 
shows that the **Monthly Income** is a main factor affecting the employee attrition, meaning the employees are underpaid.
